##### 9. Combining DataFrames in Pandas (merge, join, concat)

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Anna", "John", "Mike"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Score": [90, 75, 82]
})
```

```python
#MERGE: INNER JOIN (only matching rows)
df_inner = pd.merge(df1, df2, on="ID", how="inner")
```

```python
#MERGE: LEFT JOIN (all from left, match if possible)
df_left = pd.merge(df1, df2, on="ID", how="left")
```

```python
#MERGE: RIGHT JOIN
df_right = pd.merge(df1, df2, on="ID", how="right")
```

```python
#MERGE: OUTER JOIN (union of all keys)
df_outer = pd.merge(df1, df2, on="ID", how="outer")
```

```python
dfA = pd.DataFrame({"Age": [23, 27, 30]}, index=["Anna", "John", "Mike"])
dfB = pd.DataFrame({"Score": [90, 75, 85]}, index=["Anna", "John", "Mike"])
```

```python
# JOIN 
dfA.join(dfB)
```

```python
# Concat
df_concat = pd.concat([df1, df2], axis=0)
df_concat
```

##### 11. Using map() and apply() in Pandas

```python
df_full["Condition"] = df_full["Condition"].map({"Hipoksja": 1, "Normoksja": 0})
```

```python
df_full
```

```python
df_full["type"].unique()
```

```python
df_full["Area_normalise"] = df_full["Area"].map(lambda x: x / df_full["Area"].max())
```

```python
df_full["Round/Solidity+AR"] = df_full.apply(lambda row: (row["Round"]/ row["Solidity"])+ row["AR"], axis = 1)
```

```python
def type_to_num(type_):
    if type_ == "GL261":
        type_ = 0
        return type_ 
    
    elif type_ == 'LN229':
        type_ = 1
        return type_ 
    elif type_ == 'PANC-2':
        type_ = 2
        return type_ 
    elif type_ == "PANC-1":
        type_ = 3
        return type_ 
    elif type_ == "U87":
        type_ = 4
        return type_ 
```

```python
df_full["type_n"] = df_full["type"].apply(type_to_num)
```

```python
df_full
```
